```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Customer Churn.csv')
df.head()
```

#replacing blank with 0 as tenure is 0 and no total charges recorded

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return"no"

df['SeniorCitizen']=df["SeniorCitizen"] .apply(conv)
```

```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])
pct = ct.div(ct.sum(axis=1), axis=0) * 100
pct.plot(kind='bar', stacked=True, figsize=(4,4))
plt.title("Churn % by SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.xlabel("SeniorCitizen")
plt.legend(title="Churn")
plt.show()
```

```python
pivot = pd.crosstab(df['SeniorCitizen'], df['Churn'])
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title("Churn vs SeniorCitizen (Heatmap)")
plt.show()
```

```python
categorical_cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod',
    'Churn'
]
import seaborn as sns
import matplotlib.pyplot as plt
import math

# number of plots
n = len(categorical_cols)

# determine grid size
rows = math.ceil(n / 3)
cols = 3

plt.figure(figsize=(18, rows*4))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(rows, cols, i,)
    sns.countplot(data=df, x=col, hue="Churn")
    plt.title(col)
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```


#The countplots show how customer churn varies across different service and contract features. Categories like month-to-month contracts, lack of online security/tech support, and fiber-optic internet have noticeably higher churn levels. Features such as longer contracts and additional support services are associated with lower churn. Overall, churn patterns differ clearly across most service-related attributes.
